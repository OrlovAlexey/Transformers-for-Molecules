```python
from splitter import Splitter
from rdkit import Chem
```

```python
smiles = "O=C(O[C@@H]2Cc3c(O[C@@H]2c1ccc(O)c(O)c1)cc(O)cc3O)c4cc(O)c(O)c(O)c4" #Chem.MolFromSmiles("c1ccc(cc1)Nc2ccccc2")
mol = Chem.MolFromSmiles(smiles) 
```

```python
from rdkit.Chem import Draw
size = (500, 500)
```

```python
Draw.MolToImage(mol, size=size)
```

```python
splitter = Splitter()
list_of_smiles = splitter.get_substructures_smiles(smiles)
```

```python
list_of_smiles
```

```python
mols = []
for smiles in list_of_smiles:
    mols.append(Chem.MolFromSmiles(smiles))

Draw.MolsToImage(mols)
```
