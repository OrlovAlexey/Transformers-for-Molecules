```python
radius = 'ecfp0'
folder = 'ecfps_full'
filename = folder + '/' + radius
```

```python
# just ecfp0's in one file splitted by \n
with open('../' + filename + '.txt', 'r') as fp: 
    text = fp.read().split('\n')
```

```python
text_connected = " . ".join(text)
```

```python
text_connected[0:1000]
```

```python
from bpe_like_tokenizer import BPETokenizerSimple

tokenizer = BPETokenizerSimple()
tokenizer.train(text_connected[0:100000], vocab_size=10000, special_tokens=None)
```
